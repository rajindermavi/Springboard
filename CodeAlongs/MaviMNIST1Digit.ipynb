```python
from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml('mnist_784')
```

```python
mnist
```

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
X,y = mnist['data'],mnist['target']
```

```python
some_digit = X[1000]
some_digit_rs = some_digit.reshape(28,28)
plt.imshow(some_digit_rs, cmap = matplotlib.cm.binary, interpolation = 'nearest')
```

```python
some_digit_2 = X[4599]
some_digit_rs = some_digit_2.reshape(28,28)
plt.imshow(some_digit_rs, cmap = matplotlib.cm.binary, interpolation = 'nearest')
```

```python
x_train, x_test = X[:6000], X[6000:7000]
y_train, y_test = y[:6000], y[6000:7000]
```

```python
import numpy as np

shuffle_index = np.random.permutation(6000)
x_train, y_train = x_train[shuffle_index], y_train[shuffle_index]
```

```python
y_train_2 = (y_train.astype(np.int8) == 2)
y_test_2 = (y_test.astype(np.int8) == 2)
```

```python
y_test_2.sum()
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(tol = 0.1, max_iter = 2000)
clf.fit(x_train,y_train_2)
```

```python
clf.predict([some_digit])
```

```python
clf.predict([some_digit_2])
```

```python
from sklearn.model_selection import cross_val_score

a = cross_val_score(clf,x_train,y_train_2,cv = 3,scoring = 'accuracy' )
```

```python
a.mean()
```

```python
from sklearn import svm

clf_svm = svm.SVC()
clf_svm.fit(x_train,y_train_2)
```

```python
a = cross_val_score(clf_svm,x_train,y_train_2,cv = 3,scoring = 'accuracy' )
a.mean()
```
